```python
import os
import sys

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
from sklearn.dummy import DummyClassifier

import matplotlib.pyplot as plt

import psycopg2 as pg
```

Read the data file

```python
# The psycopg2 functions don't like numpy number types
test_df['SeniorCitizen'] = [int(test_df.loc[i,'SeniorCitizen'] == 1) for i in test_df.index]
```

```python
test_df['tenure'] = [int(test_df.loc[i,'tenure']) for i in test_df.index]
```

Split the data into features and outcome

```python
feature_df = test_df.loc[:, :'TotalCharges']
```

```python
target_df = test_df['Churn']
```

Not split off some data for final model scoring and save

```python
# Split the dataset
feature_train, feature_test, targ_train, targ_test = \
    train_test_split(feature_df, target_df, test_size=.25, random_state=42)
```

```python
# Rejoin the dataset and save the two sections to CSV file
train_df = feature_train.copy()
train_df['Churn'] = targ_train
test_df = feature_test.copy()
test_df['Churn'] = targ_test
```

```python
train_df.to_csv('../data/telecom_train.csv')
test_df.to_csv('../data/telecom_test.csv')
train_df.dtypes
```

Now split the training data into customer characteristics and account information

```python
customer_cols = ['customerID', 'gender', 'SeniorCitizen', 'Partner', 'Dependents']
account_cols = ['customerID', 'tenure', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', \
               'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',\
               'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges', 'TotalCharges']
result_cols = ['customerID', 'Churn']
```

```python
customer_df = train_df[customer_cols]
account_info_df = train_df[account_cols]
target_df = train_df[result_cols]
```

Put this data into an sql database

```python
connection_args = {
    'host': 'localhost',  # You will need to put the PUBLIC IP address of your AWS instance here
    'user': 'ubuntu',    # your username on AWS is 'ubuntu'
    'dbname': 'churn_train',    # DB that we are connecting to
    'password' : 'Metis',
    'port': 5432             # port we opened on AWS
}

# We will talk about this magic Python trick!
connection = pg.connect(**connection_args)
cur = connection.cursor()
```

```python
# Create a table if needed

type_trans = { 'object' : 'varchar(30)',
               'int64' : 'int',
               'float64' : 'float',
                'bool' : 'bit'}
type_mask = { 'object' : '%s ',
                'int64' : '%i ',
                'float' : '%f',
                'bool' : '%i'}

columns = train_df.columns
```

```python
def create_sql_table_str(name, col_names, dat_types) :
    # Paste a string together to create a table
    test_str = 'CREATE TABLE ' + name + ' ( '
    for i in range(len(col_names)-1) :
        test_str += col_names[i] + ' ' + type_trans[str(dat_types[i])] + ', '
    test_str += col_names[-1] + ' ' + type_trans[str(dat_types[-1])]
    test_str += ')'
    
    return test_str

def create_sql_row_str(frame, col_names, dat_types) :
    # Paste a string together to create a table
    test_str = ' ( '
    for i in range(len(col_names)-1) :
        test_str += col_names[i] + ', '
    test_str += col_names[-1] + ' '
    test_str += ')'
    
    return test_str
```

```python
cur.execute('begin')
```

```python
# Create a string to make a table for each of the Dataframes we've created
table_str = create_sql_table_str('customer', customer_df.columns, customer_df.dtypes)
cur.execute(table_str)

table_str = create_sql_table_str('account_info', account_info_df.columns, account_info_df.dtypes)
cur.execute(table_str)

table_str = create_sql_table_str('target_info', target_df.columns, target_df.dtypes)
cur.execute(table_str)
```

```python
def insert_data_to_sql(frame, table_name, cur) :
    # for each row in the data put the data into the database in the specified table
        # build the command string  
        base_cmd_str = 'INSERT INTO ' + table_name + create_sql_row_str(frame, frame.columns, frame.dtypes) + ' VALUES  ('

        # For each row in the dataframe insert the data
        for j in frame.index :
            cmd_str = base_cmd_str
            
            record_data = tuple(frame.loc[j, :])
        
            for obj in record_data[:-1] :
                if (type(obj) == str) | (type(obj) == bool) :
                    cmd_str += "'" + str(obj) + "', "
                else :
                    cmd_str += str(obj) + ', '
            obj = record_data[-1]
            if (type(obj) == str) | (type(obj) == bool) :
                cmd_str += "'" + str(obj) + "') "
            else :
                cmd_str += str(obj) + ') '           

            cur.execute(cmd_str, record_data)


        
```

```python
customer_df.dtypes
```

```python
# Insert the data into table
insert_data_to_sql(customer_df, 'customer', cur)
insert_data_to_sql(account_info_df, 'account_info', cur)
insert_data_to_sql(target_df, 'target_info', cur)
```
